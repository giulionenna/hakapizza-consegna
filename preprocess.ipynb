```python
def ask_llm(question):
    credentials = Credentials(
        url=ENDPOINT_URL,
        api_key=API_KEY
    )

    model = ModelInference(
        model_id="meta-llama/llama-3-3-70b-instruct",
        credentials=credentials,
        project_id=PROJECT_ID,
        params={
            "max_tokens": 20000
        }
    )

    result = model.chat(messages=[{'role': 'user', 'content': question}])
    return result['choices'][0]['message']['content']
```

```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)
    text = ""
    
    # Iterate through each page
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        text += page.get_text()
    
    return text

# Path to the PDF file
pdf_path = "Hackapizza Dataset\Codice Galattico\Codice Galattico.pdf"

# Extract text from the PDF
pdf_text = extract_text_from_pdf(pdf_path)

# Save the extracted text to a markdown file
# with open("output.md", "w", encoding="utf-8") as md_file:
#     md_file.write(pdf_text)
```

```python
prompt = "create md file with correct indentation and formatting, consider also titles and subtitles and lists. Here the text to process: \n\n"
question = prompt + pdf_text

result = ask_llm(question)
print(result)
```
